```python
# Total products shipped 
TotalProducts=pd.DataFrame(ShippingDataset.groupby(['Warehouse_block'])['Cost_of_the_Product'].count())
TotalProducts.rename(columns={'Cost_of_the_Product':'Total Products shipped'},inplace=True)
TotalProducts
```

```python
#Products has been delivered by Warehouse Blocks on time or not
pd.crosstab(index=ShippingDataset['Reached.on.Time_Y.N'],columns=ShippingDataset['Warehouse_block'])
```

```python
# MODE OF SHIPMENT

fig,ax= plt.subplots(1,3, figsize=(15,7))

color=['#9eea40','#9d9dff','#9dff9d','#ff9d9d','#ffff9d']
Shipments=pd.crosstab(ShippingDataset['Warehouse_block'],ShippingDataset['Mode_of_Shipment'])

# Creating explode data
explode = (0.0, 0.0, 0.0, 0.0, 0.2)

# Flight Shipping mode
FlightGraph=Shipments['Flight'].plot.pie(autopct='%.2f',ax=ax[0],explode=explode,colors=color,shadow=True)
FlightGraph.set_title("FLIGHT")
FlightGraph.set_ylabel(" ")

# Road Shipping mode
RoadGraph=Shipments['Road'].plot.pie(autopct='%.2f',ax=ax[1],explode=explode,colors=color,shadow=True)
RoadGraph.set_title("ROAD")
RoadGraph.set_ylabel(" ")

# Ship Shipping mode
ShipGraph=Shipments['Ship'].plot.pie(autopct='%.2f',ax=ax[2],explode=explode,colors=color,shadow=True)
ShipGraph.set_title("SHIP")
ShipGraph.set_ylabel(" ")

plt.show()
```

```python
fig,ax= plt.subplots(2,2, figsize=(15,11))


# Total products shipped 
TotalProducts=pd.DataFrame(ShippingDataset.groupby(['Warehouse_block'])['Cost_of_the_Product'].count()).plot(
    ax=ax[0,0],color=['#9dda00'],marker = 'o',ms=20,linewidth=7,mfc='#7676ff')
ax[0,0].set_title("TOTAL PRODUCTS SHIPPED" )

# WAREHOUSE BLOCKS AND TOTAL DISCOUNTS OFFERED
DiscountsGraph=pd.DataFrame(ShippingDataset.groupby('Warehouse_block') ['Discount_offered'].sum()).plot.bar(
    ax=ax[0,1], color=['#00b1b1'],edgecolor='white',linewidth=2 )
ax[0,1].set_title("TOTAL DISCOUNTS OFFERED" )


# WAREHOUSE BLOCKS AND TOTAL CUSTOMER CARE CALLS RECEIVED
CustomerCareGraph=pd.DataFrame(ShippingDataset.groupby(['Warehouse_block'])['Customer_care_calls'].sum()).plot(
    ax=ax[1,0], color=[ '#ff3b3b'], marker = 'o',ms=20,mfc='#ffe03b',linewidth=5)
ax[1,0].set_title("TOTAL CUSTOMER CARE CALLS RECEIVED")


# Products reached on time is denoted by 1 and products not reached on time is denoted by 0
ProductsDeliveryGraph=pd.crosstab(ShippingDataset['Warehouse_block'],ShippingDataset['Reached.on.Time_Y.N']).plot.bar(
    ax=ax[1,1], edgecolor='white',color=['#ff4e4e','#ffe34e'],linewidth=2)
ax[1,1].set_title(" PRODUCTS REACHED ON TIME ")

# Create a list to edit all the graphs at a time
Graphs=[TotalProducts, DiscountsGraph, CustomerCareGraph, ProductsDeliveryGraph]

# For loop- to customize all the graphs at a time
for grph in Graphs:
    grph.set_xlabel(" ")
    grph.set_ylabel(" ")
    grph.tick_params(axis='x', labelsize=12, rotation=0)
    grph.legend(fontsize=11,ncol=2)

plt.show()
```

```python
#Customer ratings 1-5 given to all the warehouse blocks
CustomerRatingGraph=pd.crosstab(ShippingDataset['Warehouse_block'],ShippingDataset['Customer_rating']).plot.bar(
    figsize= (15,5), edgecolor='white', linewidth=3, width=0.5, color=['#b1b1ff','#9d9dff','#8989ff','#7676ff','#6262ff'])
CustomerRatingGraph.set_title(" CUSTOMER CARE RATING RECEIVED BY WAREHOUSE BLOCKS ")
CustomerRatingGraph.tick_params(axis='x', labelsize=10, rotation=0)
CustomerRatingGraph.legend(ncol=5)
plt.show()
```
